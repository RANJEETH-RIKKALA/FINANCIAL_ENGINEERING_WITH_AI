## Creating CLOSE prices Dataframe for all stocks

```python
for symbol in symbols:
  df_sym = df[df['Name'] == symbol]
  # Convert the 'date' column to datetime objects before setting as index
  df_sym['date'] = pd.to_datetime(df_sym['date'])
  df_tmp = pd.DataFrame(data=df_sym['close'].to_numpy(), index=df_sym['date'], columns=[symbol])
  close_prices = close_prices.join(df_tmp) # left-join by default
```

## Handling Missing Data

```python
close_prices['IBM'].plot()
```

```python
# drops rows with all nan
# most likely correspond to weekends, holidays (non-trading days)
close_prices.dropna(axis=0, how='all', inplace=True)
```

```python
close_prices['IBM'].plot()
```

```python
# Check Missing Values Count
close_prices.isna().sum()
```

```python
close_prices.isna().sum().sum()
```

```python
# Check how many stocks have missing value in the first row of Dataframe
close_prices.iloc[0,:].isna().sum()
```

### Forward Fill NA values

```python
close_prices.fillna(method='ffill', inplace=True)
```

```python
close_prices.isna().sum().sum()
```

### Backward Fill NA values

```python
close_prices.bfill(inplace=True)
```

```python
close_prices.isna().sum().sum()
```

### Plot the remaining stock close prices

```python
close_prices.plot(legend=False,figsize=(10,10));
```

### Normalized Close Prices

```python
close_prices_normalized = close_prices / close_prices.iloc[0]
```

```python
close_prices_normalized.plot(legend=False,figsize=(10,10));
```

```python
# finding the STOCK that has gone up 20X compared to initial Value
close_prices_normalized.max().sort_values(ascending=False).head(1)
```
